## Load Data

### Subtask:
Load the 'Zomato_Cleaned_and_Engineered.csv' dataset into a pandas DataFrame and set the dark background style for plots.


**Reasoning**:
First, I'll import the necessary libraries: pandas for data manipulation and matplotlib.pyplot for plotting and setting the plot style. Then, I'll load the dataset into a DataFrame and set the plotting style to 'dark_background'.



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Zomato_Cleaned_and_Engineered.csv')
plt.style.use('dark_background')

print("Dataset loaded successfully and plotting style set to 'dark_background'.")
print(df.head())
```

## Delivery Time by City (Box Plot)

### Subtask:
Generate a box plot to visualize the distribution of `Time_taken(min)` across different `City` values.


**Reasoning**:
I will generate a box plot to visualize the distribution of delivery time by city, following all the instructions provided, including importing seaborn, setting plot aesthetics, and saving the figure.



```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='City', y='Time_taken (min)', data=df, palette='viridis')

plt.title('Delivery Time Distribution by City', fontsize=16, color='white', fontweight='bold')
plt.xlabel('City', fontsize=12, color='white')
plt.ylabel('Delivery Time (minutes)', fontsize=12, color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('delivery_time_by_city_boxplot.png', dpi=150)
plt.show()

print("Box plot of Delivery Time by City generated and saved as 'delivery_time_by_city_boxplot.png'.")
```

**Reasoning**:
The previous code produced a `FutureWarning` regarding the `palette` argument without `hue`. To fix this and follow the recommended practice for `seaborn` versions 0.14.0 and above, I will explicitly set `hue='City'` and `legend=False` in the `sns.boxplot` function. I also noticed that the instructions mentioned using `plt.subplots()` and `ax.set_title()` etc., which was not fully followed in the previous code. I will update the code to use the `ax` object for setting plot properties to ensure cleaner code and better control over the plot.



```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(x='City', y='Time_taken (min)', data=df, palette='viridis', hue='City', legend=False, ax=ax)

ax.set_title('Delivery Time Distribution by City', fontsize=16, color='white', fontweight='bold')
ax.set_xlabel('City', fontsize=12, color='white')
ax.set_ylabel('Delivery Time (minutes)', fontsize=12, color='white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('delivery_time_by_city_boxplot.png', dpi=150)
plt.show()

print("Box plot of Delivery Time by City generated and saved as 'delivery_time_by_city_boxplot.png'.")
```

## Order Hour vs. Order Day of Week Heatmap

### Subtask:
Create a pivot table with `Order_hour` as index, `Order_day_of_week` as columns, and the mean of `Time_taken(min)` as values. Visualize this pivot table as a heatmap. Ensure the heatmap has a bold white title, figsize=(12,6), white axes labels, and use a vibrant sequential color palette. Save the plot as 'order_time_heatmap.png' (dpi=150, tight_layout()).


**Reasoning**:
I will create a pivot table to aggregate the average delivery time by order hour and day of week, then visualize it as a heatmap with specified aesthetics and save the plot, following all the instructions provided.



```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create a pivot table
heatmap_data = df.pivot_table(index='Order_hour', columns='Order_day_of_week', values='Time_taken (min)', aggfunc='mean')

# 2. Create a figure and an axes object
fig, ax = plt.subplots(figsize=(12, 6))

# 3. Generate a heatmap
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt=".1f", linewidths=.5, ax=ax)

# 4. Set the title
ax.set_title('Average Delivery Time by Order Hour and Day of Week', fontsize=16, color='white', fontweight='bold')

# 5. Label the x-axis and y-axis
ax.set_xlabel('Order Day of Week', fontsize=12, color='white')
ax.set_ylabel('Order Hour', fontsize=12, color='white')

# 6. Set the color of the x and y axis ticks
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# 7. Use plt.tight_layout()
plt.tight_layout()

# 8. Save the plot
plt.savefig('order_time_heatmap.png', dpi=150)

# 9. Display the plot
plt.show()

# 10. Print a confirmation message
print("Heatmap of Average Delivery Time by Order Hour and Day of Week generated and saved as 'order_time_heatmap.png'.")
```

## Average Delivery Time by Weather and Traffic (Grouped Bar Chart)

### Subtask:
Create a grouped bar chart showing the average `Time_taken(min)` grouped by `Weather_conditions` and `Road_traffic_density`. Include value annotations.


**Reasoning**:
I will calculate the average delivery time grouped by weather conditions and road traffic density, then create a grouped bar chart with value annotations, adhering to all specified plot aesthetics and saving instructions.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Calculate the average Time_taken (min) grouped by Weather_conditions and Road_traffic_density
avg_time_by_weather_traffic = df.groupby(['Weather_conditions', 'Road_traffic_density'])['Time_taken (min)'].mean().reset_index()

# 2. Create a figure and an axes object
fig, ax = plt.subplots(figsize=(12, 6))

# 3. Generate a grouped bar chart
sns.barplot(x='Weather_conditions', y='Time_taken (min)', hue='Road_traffic_density', data=avg_time_by_weather_traffic, palette='viridis', ax=ax)

# 4. Set the title
ax.set_title('Average Delivery Time by Weather Condition and Traffic Density', fontsize=16, color='white', fontweight='bold')

# 5. Label the x-axis and y-axis
ax.set_xlabel('Weather Condition', fontsize=12, color='white')
ax.set_ylabel('Average Delivery Time (minutes)', fontsize=12, color='white')

# 6. Set the color of the x and y axis ticks
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# 7. Add gridlines
ax.grid(alpha=0.3)

# 8. Adjust the plot layout
plt.tight_layout()

# 9. Add value annotations on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', fontsize=8, color='white')

# 10. Save the plot
plt.savefig('avg_time_weather_traffic_barplot.png', dpi=150)

# 11. Display the plot
plt.show()

# 12. Print a confirmation message
print("Grouped bar chart of Average Delivery Time by Weather Condition and Traffic Density generated and saved as 'avg_time_weather_traffic_barplot.png'.")
```
